# WHAT IS LINEAR REGRESSION IN MACHINE LEARNING ?

## Linear regression is a fundamental and widely used statistical technique in machine learning and data analysis.
## It is a type of supervised learning algorithm used for predicting continuous output values based on input features.
## In simple terms, it helps establish a linear relationship between the independent variables (input features) and the dependent variable (output).
## The goal of linear regression is to find the best-fitting straight line (also known as the regression line) that minimizes the difference between the predicted values and the actual target values.


## This line is represented by the equation:
## y = mx + b
## Where:
## y is the predicted output (dependent variable).
## x is the input feature (independent variable).
## m is the slope (also called the coefficient), which represents the change in the output concerning a change in the input.
## b is the y-intercept, which is the value of y when x is 0.

## In more complex cases with multiple input features, the equation becomes a bit more intricate:
## y = b + m1x1 + m2x2 + ... + mn*xn
## Here, m1, m2, ..., mn are the coefficients for each respective input feature x1, x2, ..., xn.


## The process of linear regression involves estimating the values of the coefficients (m and b) based on the input data.
## The most common method used for this estimation is the Ordinary Least Squares (OLS) algorithm, which minimizes the sum of the squared differences between the predicted and actual values.


# ABOUT THE HOUSE PRICE PREDICTION DATASET :-

## The features (also known as variables or attributes) in a House Price Prediction dataset are the characteristics or factors that are used to predict the target variable, which is the house price.
## The specific features may vary depending on the dataset and the data sources used. However, here are some common features typically found in a House Price Prediction dataset:

## 1. Living Area: The total living area of the house in square feet or square meters.
## 2. Number of Bedrooms: The total number of bedrooms in the house.
## 3. Number of Bathrooms: The total number of bathrooms in the house, including full and half bathrooms.
## 4. Lot Size: The size of the land on which the house is built, usually measured in square feet or acres.
## 5. Location: Geographic information such as city, neighborhood, and zip code, which can have a significant impact on house prices.
## 6. Latitude and Longitude: The exact geographical coordinates of the property.
## 7. House Age: The age of the house in years since construction or renovation.
## 8. Property Condition: An assessment of the overall condition of the property (e.g., excellent, good, fair, poor).
## 9. Amenities: Features related to additional amenities and facilities such as a garage, swimming pool, garden, fireplace, etc.
## 10. Nearby Services: Proximity to essential services like schools, hospitals, public transportation, shopping centers, and parks.
## 11. Property Type: The type of property, such as single-family home, condominium, townhouse, etc.
## 12. Number of Floors: The total number of floors or stories in the house.
## 13. Heating and Cooling Systems: Information about the heating and cooling systems used in the property.
## 14. Energy Efficiency: Data related to energy-efficient features, such as solar panels or energy-efficient appliances.
## 15. Historical Price Data: In some cases, historical price data for the property, such as previous sale prices, might be available.
## 16. Market Trends: Information about current market trends and historical changes in house prices.
## 17. Property Tax: The amount of property tax associated with the property.
## 18. Accessibility: Information about accessibility features, especially for differently-abled individuals.
## 19. Crime Rate: Data related to the crime rate in the neighborhood or surrounding areas.
## 20. Views and Scenery: Whether the property offers attractive views or scenic surroundings.

## Feature Extraction :-

```python
X = house_df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition']]
y = house_df['price']
```

## Splitting Dataset :-

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.30 , random_state = 101)
```

## Fitting the Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
hpp = LinearRegression()
```

```python
hpp.fit(X_train , y_train)
```

```python
# Coefficients of the linear regression model
coeff_df = pd.DataFrame(hpp.coef_ , X.columns , columns = ['Coefficient'])
coeff_df
```

## Metrics

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
hpp_predict = hpp.predict(X_test)
```

```python
hpp_mse = mean_squared_error(y_test , hpp_predict)
```

```python
hpp_r2 = r2_score(y_test , hpp_predict)
```

```python
print("Mean Squared Error : " , hpp_mse)
print("R-Sqaured : " , hpp_r2)
```

## Visualization

```python
plt.scatter(y_test, hpp_predict)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs. Predicted Prices")
plt.show()
```

```python
# Residual plot
residuals = y_test - hpp_predict
plt.scatter(y_test, residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel("Actual Prices")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```
